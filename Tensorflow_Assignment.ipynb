```python
import pandas as pd
```

```python
data = pd.read_csv('hour.csv')
```

```python
data = data.drop(['instant','dteday','casual' , 'registered'],axis=1)
```

```python
X = data.drop('cnt',axis=1)
y = data['cnt']

from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X.values,y.values,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf
```

```python
models = tf.keras.models.Sequential()
```

```python
models.add(tf.keras.layers.Dense(input_dim=12,units=32,activation='relu', kernel_initializer='normal'))
models.add(tf.keras.layers.Dense(units=64,activation='relu',kernel_initializer='normal'))
models.add(tf.keras.layers.Dense(units=1,activation='linear',kernel_initializer='normal'))

```

```python
models.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mean_absolute_error'])
```

```python
models.summary()
```

```python
models.fit(X_train,y_train,nb_epoch=100,batch_size=10)
```

```python
y_pred = models.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test,y_pred))
```
